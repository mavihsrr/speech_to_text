```python
import tkinter as tk
from tkinter import filedialog
from tkinter import ttk
import re
from datetime import datetime

RECENT_FILES_LIMIT = 5  # Maximum number of recent files to keep

def extract_lines():
    user_condition = condition_entry.get()
    pattern = rf'(\w+ \d+ \d+:\d+:\d+ .*{re.escape(user_condition)}.*)'

    file_path = filedialog.askopenfilename(filetypes=[('Text Files', '*.txt')])

    if file_path:
        with open(file_path, 'r', encoding='utf-8', errors='ignore') as file:
            data = file.read()
            matches = re.findall(pattern, data)
            matches = sorted(matches, key=lambda x: extract_timestamp(x))

            treeview.delete(*treeview.get_children())

            start_time_str = start_time_entry.get()
            end_time_str = end_time_entry.get()

            start_time = datetime.strptime(start_time_str, '%b %d %H:%M:%S')
            end_time = datetime.strptime(end_time_str, '%b %d %H:%M:%S')

            for match in matches:
                timestamp = extract_timestamp(match)
                line = match.replace(timestamp, '', 1).strip()
                line_time = datetime.strptime(timestamp, '%b %d %H:%M:%S')
                if start_time <= line_time <= end_time:
                    treeview.insert('', tk.END, values=(timestamp, line))

            count_label.config(text=f"Matches Found: {len(treeview.get_children())}")

            file_name_label.config(text=f"Selected File: {file_path}")

            add_to_recent_files(file_path)

def extract_timestamp(line):
    timestamp = re.search(r'^(\w+ \d+ \d+:\d+:\d+)', line)
    if timestamp:
        return timestamp.group(1)
    return ''

def add_to_recent_files(file_path):
    recent_files = get_recent_files()
    if file_path in recent_files:
        recent_files.remove(file_path)
    recent_files.insert(0, file_path)
    if len(recent_files) > RECENT_FILES_LIMIT:
        recent_files = recent_files[:RECENT_FILES_LIMIT]
    save_recent_files(recent_files)
    update_recent_files_menu()

def get_recent_files():
    try:
        with open("recent_files.txt", "r") as file:
            recent_files = file.read().splitlines()
        return recent_files
    except FileNotFoundError:
        return []

def save_recent_files(recent_files):
    with open("recent_files.txt", "w") as file:
        file.write('\n'.join(recent_files))

def open_recent_file(file_path):
    condition_entry.delete(0, tk.END)
    condition_entry.insert(tk.END, "")
    start_time_entry.delete(0, tk.END)
    start_time_entry.insert(tk.END, "")
    end_time_entry.delete(0, tk.END)
    end_time_entry.insert(tk.END, "")
    extract_lines()

def update_recent_files_menu():
    recent_files = get_recent_files()
    recent_files_menu.delete(0, tk.END)
    for i, file_path in enumerate(recent_files):
        recent_files_menu.add_command(label=file_path, command=lambda f=file_path: open_recent_file(f))

window = tk.Tk()
window.title("LogExtractor")


condition_label = tk.Label(window, text="Enter the condition:")
condition_label.pack()
condition_entry = tk.Entry(window)
condition_entry.pack()

start_time_label = tk.Label(window, text="Start Timestamp:")
start_time_label.pack()
start_time_entry = tk.Entry(window)
start_time_entry.pack()

end_time_label = tk.Label(window, text="End Timestamp:")
end_time_label.pack()
end_time_entry = tk.Entry(window)
end_time_entry.pack()

select_file_button = tk.Button(window, text="Select File", command=extract_lines)
select_file_button.pack()

file_name_label = tk.Label(window, text="Selected File: ")
file_name_label.pack()

recent_files_menu = tk.Menu(window, tearoff=0)
update_recent_files_menu()
recent_files_menu_label = tk.Label(window, text="Recent Files:")
recent_files_menu_label.pack()
recent_files_menu_button = tk.Menubutton(window, text="Open Recent", relief=tk.RAISED)
recent_files_menu_button.pack()
recent_files_menu_button.config(menu=recent_files_menu)

content_frame = tk.Frame(window)
content_frame.pack(fill=tk.BOTH, expand=True)

treeview = ttk.Treeview(content_frame, columns=('Timestamp', 'Line'), show='headings')
treeview.heading('Timestamp', text='Timestamp')
treeview.heading('Line', text='Line')

treeview.column('Timestamp', width=200)
treeview.column('Line', width=800)

scrollbar = ttk.Scrollbar(content_frame, orient=tk.VERTICAL, command=treeview.yview)
treeview.configure(yscroll=scrollbar.set)

treeview.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
scrollbar.pack(side=tk.RIGHT, fill=tk.Y)

count_label = tk.Label(window, text="Matches Found: 0")
count_label.pack()

window.mainloop()
```
