```python
import re
```

```python
# User input for condition
user_condition = input("Enter the condition to match: ")

# regex pattern for both timestamp and full line of log
pattern = rf'(\w+ \d+ \d+:\d+:\d+ .*{re.escape(user_condition)}.*)'
```

```python
with open('messages.txt', 'r', encoding='utf-8') as file:
    data = file.read()

    matches = re.findall(pattern, data)

    for match in matches:
        print(match)
        
```
